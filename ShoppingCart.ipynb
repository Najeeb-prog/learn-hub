```python
electronics = {
    "mobile": 55000,
    "washing machine": 42000,
    "ac": 67000
}
wellness = {
    "face wash": 950,
    "serum": 899,
    "sun screen": 1600
}
nutrition = {
    "whey": 5200,
    "creatine": 1200,
    "fish oil": 950
}

total_amount = 0
selected_items = []


def view_total_amount():
    print(f"The total amount is: {total_amount}\n")


def view_cart():
    if not selected_items:
        print("Your cart is empty.\n")
    else:
        print("Items in your cart:")
        print("\n".join(selected_items))
        print()


def checkout_main():
    if total_amount > 0:        
        print("\nChecking out:")
        print("\n".join(selected_items))
        print(f"\nTOTAL: {total_amount}\n")
    else:
        print("\nYou didn't make a purchase! Please select items to buy first.\n")


print("Welcome to Shopping!\n")
while True:
    print("1. Electronics\n2. Wellness\n3. Nutrition\n4. View Cart\n5. View Total\n00. Exit\n01. Checkout\n")
    category_selection = input("Enter the category (1/2/3/4/5/00/01): \n")
    
    if category_selection == '00':
        print("Quitting...")
        break

    elif category_selection == '1':
        while True:
            print("You chose Electronics:\n")
            for key, value in electronics.items():
                print(f"{key} - {value}")
            print("Type 'break' to quit")
            buy = input("\nEnter the item you want to buy: ").lower()
            if buy == "break":
                break
            if buy in electronics:
                selected_items.append(buy)
                total_amount += electronics[buy]
                print(f"Added {buy} to cart.\n")
            else:
                print("Item not found!\n")

    elif category_selection == '2':
        while True:
            print("You chose Wellness:\n")
            for key, value in wellness.items():
                print(f"{key} - {value}")
            print("Type 'break' to quit")
            buy = input("\nEnter the item you want to buy: ").lower()
            if buy == "break":
                break
            if buy in wellness:
                selected_items.append(buy)
                total_amount += wellness[buy]
                print(f"Added {buy} to cart.\n")
            else:
                print("Item not found!\n")

    elif category_selection == "3":
        while True:
            print("You chose Nutrition:\n")
            for key, value in nutrition.items():
                print(f"{key} - {value}")
            print("Type 'break' to quit")
            buy = input("\nEnter the item you want to buy: ").lower()
            if buy == "break":
                break
            if buy in nutrition:
                selected_items.append(buy)
                total_amount += nutrition[buy]
                print(f"Added {buy} to cart.\n")
            else:
                print("Item not found!\n")

    elif category_selection == '4':
        view_cart()

    elif category_selection == '5':
        view_total_amount()

    elif category_selection == '01':
        checkout_main()
```
